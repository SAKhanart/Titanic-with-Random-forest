```python
import pandas as pd
import numpy as np
```

```python
data_train=pd.read_csv("/content/train.csv")
data_test=pd.read_csv("/content/test.csv")

data_train.head()
```

```python
data_test.head()
```

```python
surv_train= data_train['Survived']
```

```python
surv_train.head()
```

```python
data_train.head()
```

```python
data_train=data_train.drop(['Survived','Name','Ticket','Cabin','Fare','PassengerId','SibSp','Parch'],axis=1)  #axis =1 meand it refers colomn instead of Rows
passengerId=data_test['PassengerId']
data_test=data_test.drop(['Name','Ticket','Cabin','Fare','SibSp','Parch','PassengerId'],axis=1)
```

```python
passengerId.head()
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
data_train.info()
data_test.info()
```

```python
# Alternative: Use median instead of mean
median_age_train = data_train['Age'].median()
data_train['Age'].fillna(median_age_train, inplace=True) #inplace=true means it does not need to another varrible, it will modigy the orginal dataframe

median_age_test = data_test['Age'].median()
data_test['Age'].fillna(median_age_test, inplace=True)
```

```python
data_train.info()
data_test.info()
```

```python
# Convert the 'Sex' column from string (male/female) to integer (0/1)
# We'll map 'male' to 0 and 'female' to 1.
data_train['Sex'] = data_train['Sex'].map({'male': 0, 'female': 1})
data_test['Sex'] = data_test['Sex'].map({'male': 0, 'female': 1})

# Now you can safely convert the column to an integer type
data_train['Sex'] = data_train['Sex'].astype(int)
data_test['Sex'] = data_test['Sex'].astype(int)

# You can verify the changes
data_train.info()
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
data_train['Embarked'].unique() #it show the data catagory
```

```python
mode_embarked = data_train['Embarked'].mode()[0]
data_train['Embarked'].fillna(mode_embarked, inplace=True)
```

```python
data_train['Embarked'] = data_train['Embarked'].map({'S': 0, 'C': 1, 'Q':2,})
data_train['Embarked'] = data_train['Embarked'].astype(int)
data_train.info()
data_test.head()
```

```python
data_test['Embarked'] = data_test['Embarked'].map({'S': 0, 'C': 1, 'Q':2,})
data_test['Embarked'] = data_test['Embarked'].astype(int)
data_test.info()
data_test.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Create and fit scaler
scaler = StandardScaler()
# Apply to numerical columns only
data_train['Age'] = scaler.fit_transform(data_train[['Age']])
data_test['Age']= scaler.fit_transform(data_test[['Age']])
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
#Randomforest
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=300, max_depth=3, random_state=42)
model.fit(data_train,surv_train)
```

```python
pred=model.predict(data_test)
```

```python
sub_rep = pd.DataFrame({'PassengerId': passengerId, 'Survived': pred})
sub_rep.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

```python
#DecisionTree
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(min_impurity_decrease=0.001,min_samples_leaf=10)
dt_model.fit(data_train,surv_train)
pred2=dt_model.predict(data_test)
```

```python
sub_rep2 = pd.DataFrame({'PassengerId': passengerId, 'Survived': pred2})
sub_rep2.to_csv('submission2.csv', index=False)
print("Your submission was successfully saved!")
```
